```python
# A class named Child is created so that multiple objects can be created on order to store data of multiple students.
class Child:
    # Create object of Child class to declare and initialize name, age, gender, height and weight fields for the object.
    # This constructor also stores the pronoun of the child.
    # It also has a dictionary named food_and_calorie which has the data of calorie per gram in the respective food items.
    def __init__(self, name, age, gender, height, weight):
        # Initialize general child's data
        self.name = name
        self.age = age
        self.gender = gender
        self.height = height
        self.weight = weight
        
        # Store the child's pronoun
        self.pronoun = "him" if self.gender.lower() == "male" else "her"
        self.pronoun2 = "He" if self.gender.lower() == "male" else "She"
        
        # Dictionary for calorie per gram of each food item
        self.food_and_calorie = {
            "milk":100/100,
            "egg":155/100,
            "rice":130/100,
            "lentils":113/100,
            "vegetable":85/100,
            "meat":143/100
        }
        
        # Stores required calorie by the child according to his/her age.
        self.required_calories = 800 if self.age in range(0, 2, 1) else 1400 if self.age in range(2, 4, 1) else 1800 if self.age in range(4, 8, 1) else 0

    # A funtion which returns a string of statement about child's BMI and nourishment.
    def get_required_calorie_and_bmi(self):
        # Claculation of BMI
        BMI = round(self.weight/(self.height**2)*703,2)
        
        # Store nourishmment of child in a variable according to BMI range.
        if BMI < 16:
            nourishment = "Severly Underweight"
        elif 16 <= BMI < 18.5:
            nourishment = "Underweight"
        elif 18.5 <= BMI < 25:
            nourishment = "Healthy"
        elif 25 <= BMI < 30:
            nourishment = "Overweight"
        elif BMI >= 30:
            nourishment = "Obese"
        else:
            nourishment = "unknown"
        
        # return the string of information which has chil's name, age, BMI, required calories and nourishment.
        return f"As {self.name}'s age is {self.age} and BMI is {BMI} the minimum calorie required by {self.pronoun} is: {self.required_calories}. {self.pronoun2} is {nourishment}\n"
    
    # A function which returns and array [calorie consumed today by the child, is the child undernourished today?]
    # It takes a parameter : dictionary containing information of each food item and it's weight in grams eaten today by the child.
    def get_daily_calorie_consumption(self, foods):
        # default value of daily calorie consumed is set to 0.
        daily_calorie = 0
        # iterate to each item of the foods dictionary provided by the user, which contains the weight of each food item consumed bby the child.
        # Add each food's calorie in daily_calorie varable
        for i in foods:
            food = i.lower()
            quantity = foods[i]
            daily_calorie += self.food_and_calorie[food]* quantity
            
        # Check if the calorie taken by the child tofay is less than the required calorie or not
        # if yes then set today_undernurished_bool to empty string. else set it to NOT.
        if daily_calorie < self.required_calories:
            today_undernurished_bool = ""
        else:
            today_undernurished_bool = "NOT "
            
        # return an array with 2 elemnts. first element:calories consumed today, second element: whether the child is under nourished or not.
        return [round(daily_calorie,2),today_undernurished_bool]
            
            
                
        
```

```python
# Initialize an empty aray to store objects of Child class.
list_of_children = []
```

```python
# This main function is used to display menu and get user input and display respective output.
def main():
    # Show menu until users choice is between 1 to 5 else exit menu.
    while True:
        # Menu
        print("\n--------------------------------------\n")
        print("Enter the serial number to select the option")
        print("1. To add new child's details.")
        print("2. To show list of children.")
        print("3. To get required calorie of the child.")
        print("4. To get child's daily calorie and that the child is under nourished or not.")
        print("5. To delete child's data.")
        print("Enter any other option to exit.")
        
        # accept user's choice as an integer
        choice = int(input(">>> "))
        # just for decoration and spacing
        print("\n--------------------------------------\n")
        
        # If choice is 1: input child's name, age, gender, weight in pounds, and height in inches. create an object of Child class and sotre the object in list_of_children list.
        # also check for valid range and input format of the data, if not in required format/range then show a message and ask for the input again.
        if(choice == 1):
            name = input("Enter name: ")
            while(True):
                age = int(input("Enter age: "))
                if age in list(range(0,9,1)):
                    break
                else:
                    print("Age should be in between 0-8 years only.")
            while(True):
                gender = input("Enter gender (male/female): ")
                if gender == "male" or gender == "female":
                    break
                else:
                    print("Gender should be eigther male or female.")
            while True:
                weight = float(input("Enter weight in pounds: "))
                if weight >0:
                    break
                else:
                    print("Weight should be a positive number.")
            while True:
                height = float(input("Enter height in inches: "))
                if height>0:
                    break
                else:
                    print("Height should be a positive number.")
            list_of_children.append(Child(name, age, gender, height, weight))
            print(f"Details of {name} added successfully.")
            
        # if user's choice is 2: show the  details of each child stored in the list_of_children list.
        elif(choice == 2):
            if len(list_of_children) == 0:
                print("No child details found! Please add one first.")
            else:
                for i in list_of_children:
                    print("name:", i.name, "age:", i.age, "gender:", i.gender, "height:", i.height, "weight:", i.weight)
        
        # If user's choice is 3: 
        # If list_of_children has only one item then show the caloire and bmi of the child
        # If list_of_children has more than one item then ask for name of the child
        # If the child's name mathces with any of the name is the list_of_children list then show it's BMI and Calorie else show an eror message.
        elif(choice == 3):
            if len(list_of_children) == 0:
                print("No child details found! Please add one first.")
            elif(len(list_of_children) == 1):
                print(list_of_children[0].get_required_calorie_and_bmi())
            else:
                name = input("Enter name of the child to get details of: ")
                for i in list_of_children:
                    if i.name.lower() == name.lower():
                        print(i.get_required_calorie_and_bmi())
                        break
                else:
                    print(f"No child details found with name {name}! Please add one first.")
                    
        # If user's choice is 4:
        # If list_of_children has only one item then select the child
        # If list_of_children has more than one item then ask for child's name, if the name is present in the list_of_children list then select that child else show error message
        # After selecting the child ask for food taken by the child today in grams and store it in a dictionary.
        # Call the function get_daily_calorie_consumption of Child class with the parameter (food_taken_today_dictionary).
        elif(choice == 4):
            if len(list_of_children) == 0:
                print("No child details found! Please add one first.")
                continue
            elif(len(list_of_children) == 1):
                child = list_of_children[0]
            else:
                name = input("Enter name of the child to get details of: ").lower()
                for i in list_of_children:
                    if i.name.lower() == name:
                        child = i
                        break
                else:
                    print(f"No child details found with name {name}! Please add one first.")
                    continue
                        
            print("Enter details of food eaten by the child (in grams).")
            food = {}
            while True:
                food["milk"] = float(input("Milk: "))
                if food["milk"] >=0 :
                    break
                else:
                    print("Food items quantity should be >= 0.")
            while True:
                food["egg"] = float(input("Egg: "))
                if food["egg"] >=0 :
                    break
                else:
                    print("Food items quantity should be >= 0.")
            while True:
                food["rice"] = float(input("Rice: "))
                if food["rice"] >=0 :
                    break
                else:
                    print("Food items quantity should be >= 0.")
            while True:
                food["lentils"] = float(input("lentils: "))
                if food["lentils"] >=0 :
                    break
                else:
                    print("Food items quantity should be >= 0.")
            while True:
                food["vegetable"] = float(input("Vegetable: "))
                if food["vegetable"] >=0 :
                    break
                else:
                    print("Food items quantity should be >= 0.")
            while True:
                food["meat"] = float(input("Meat: "))
                if food["meat"] >=0 :
                    break
                else:
                    print("Food items quantity should be >= 0.")

            result = child.get_daily_calorie_consumption(food)
            print(f"Calorie taken by {child.name} today is {result[0]} and {child.pronoun2.lower()} is {result[1]}undernourished.")
                
        # If user's choice is 5:
        # If list_of_children has only one element then delete that child's data.
        # if list_of_children has more than one element than ask for name. If the name is present in the list_of_children list then delete that child's data.
        elif(choice == 5):
            name = ""
            if len(list_of_children) == 0:
                print("No child details found! Please add one first.")
                continue
            elif(len(list_of_children) == 1):
                child = list_of_children[0]
                name = child.name
            else:
                name = input("Enter name of the child to delete details of: ").lower()
                for i in list_of_children:
                    if i.name.lower() == name.lower():
                        child = i
                        break
                else:
                    print(f"No child details found with name {name}! Please add one first.")
                    continue
            print(f"Are you sure you want to delete {name}'s data?")
            ch = input("Y/N >>> ")
            if(ch.lower() == "y"):
                list_of_children.remove(child)
                print(f"{name}'s data deleted successfully!")
            else:
                print("Aborted!")
        
        
        # If the user enters any other number then exit the loop.
        else:
            break
```

```python
# Call this main function to start the program
main()
```
