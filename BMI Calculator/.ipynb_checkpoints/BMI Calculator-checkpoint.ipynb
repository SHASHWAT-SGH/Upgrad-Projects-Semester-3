```python
class Child:
    
    def __init__(self, name, age, gender, height, weight):
        self.name = name
        self.age = age
        self.gender = gender
        self.height = height
        self.weight = weight
        
        self.pronoun = "him" if self.gender.lower() == "male" else "her"
        
        self.food_and_calorie = {
            "milk":100/100,
            "egg":155/100,
            "rice":130/100,
            "lentils":113/100,
            "vegetable":85/100,
            "meat":143/100
        }
        
        self.required_calories = 800 if self.age in range(0, 2, 1) else 1400 if self.age in range(2, 4, 1) else 1800 if self.age in range(4, 8, 1) else 0
        print(f"As {self.name}'s age is {self.age} the minimum calorie required by {self.pronoun} is: {self.required_calories}\n")
    
    def get_required_calorie(self):
        return f"As {self.name}'s age is {self.age} the minimum calorie required by {self.pronoun} is: {self.required_calories}\n"
    
    def get_daily_calorie_consumption(self, foods):
        daily_calorie = 0
        for i in foods:
            food = i.lower()
            quantity = foods[i]
            daily_calorie += self.food_and_calorie[food]* quantity
            
        BMI = self.weight/(self.height**2)*703
        
        if BMI < 16:
            nourishment = "Severly Underweight"
        elif 16 <= BMI < 18.5:
            nourishment = "Underweight"
        elif 18.5 <= BMI < 25:
            nourishment = "Healthy"
        elif 25 <= BMI < 30:
            nourishment = "Overweight"
        elif BMI >= 30:
            nourishment = "Obese"
        else:
            nourishment = "unknown"
            
        return [daily_calorie, nourishment]
            
            
                
        
```

```python
list_of_children = []
```

```python
def main():
    while True:
        print("\n--------------------------------------\n")
        print("Enter the serial number to select the option")
        print("1. To add new child's details.")
        print("2. To show list of children.")
        print("3. To get required calorie by the child.")
        print("4. To get child's daily calorie and that the child is under nourished or not.")
        print("5. To delete child's data.")
        print("Enter any other option to exit.")
        
        choice = int(input(">>> "))
        print("\n--------------------------------------\n")
        
        if(choice == 1):
            name = input("Enter name: ")
            while(True):
                age = int(input("Enter age: "))
                if age in list(range(0,9,1)):
                    break
                else:
                    print("Age should be in between 0-8 years only.")
            while(True):
                gender = input("Enter gender (male/female): ")
                if gender == "male" or gender == "female":
                    break
                else:
                    print("Gender should be eigther male or female.")
            weight = float(input("Enter weight in pounds: "))
            height = float(input("Enter height in inches: "))
            list_of_children.append(Child(name, age, gender, height, weight))
            print(f"Details of {name} added successfully.")
            
        elif(choice == 2):
            if len(list_of_children) == 0:
                print("No child details found! Please add one first.")
            else:
                for i in list_of_children:
                    print("name:", i.name, "age:", i.age, "gender:", i.gender, "height:", i.height, "weight:", i.weight)
        
        elif(choice == 3):
            if len(list_of_children) == 0:
                print("No child details found! Please add one first.")
            elif(len(list_of_children) == 1):
                print(list_of_children[0].get_required_calorie())
            else:
                name = input("Enter name of the child to get details of: ")
                for i in list_of_children:
                    if i.name.lower() == name.lower():
                        print(i.get_required_calorie())
                        break
                else:
                    print(f"No child details found with name {name}! Please add one first.")
                    
        elif(choice == 4):
            if len(list_of_children) == 0:
                print("No child details found! Please add one first.")
                continue
            elif(len(list_of_children) == 1):
                child = list_of_children[0]
            else:
                name = input("Enter name of the child to get details of: ").lower()
                for i in list_of_children:
                    if i.name.lower() == name:
                        child = i
                        break
                else:
                    print(f"No child details found with name {name}! Please add one first.")
                    continue
                        
            print("Enter details of food eaten by the child (in grams).")
            food = {}
            food["milk"] = float(input("Milk: "))
            food["egg"] = float(input("Egg: "))
            food["rice"] = float(input("Rice: "))
            food["lentils"] = float(input("lentils: "))
            food["vegetable"] = float(input("Vegetable: "))
            food["meat"] = float(input("Meat: "))
            result = child.get_daily_calorie_consumption(food)
            print(f"Calorie taken by {child.name} today is {result[0]} : {result[1]}")
                
        elif(choice == 5):
            name = ""
            if len(list_of_children) == 0:
                print("No child details found! Please add one first.")
                continue
            elif(len(list_of_children) == 1):
                child = list_of_children[0]
                name = child.name
            else:
                name = input("Enter name of the child to delete details of: ").lower()
                for i in list_of_children:
                    if i.name.lower() == name.lower():
                        child = i
                        break
                else:
                    print(f"No child details found with name {name}! Please add one first.")
                    continue
            print(f"Are you sure you want to delete {name}'s data?")
            ch = input("Y/N >>> ")
            if(ch.lower() == "y"):
                list_of_children.remove(child)
                print(f"{name}'s data deleted successfully!")
            else:
                print("Aborted!")
        
        else:
            break
```

```python
main()
```
